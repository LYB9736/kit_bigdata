1-1 빅데이터의 3V에 대해 설명하시오.

3V 
1. 거대한 규모 (Volume) : 방대한 양의 데이터(테라, 페타바이트 이상)
2. 빠른 속도(Velocity) : 실시간 생성, 빠른 속도로 데이터를 처리 및 분석
3. 높은 다양성(Variety) : 정형(DBMS  등) 및 비정형(영상, 음성, 텍스트 SNS 등)

기존의 데이터베이스 관리 도구, 관리 시스템의 능력을 넘어 대량의 정형, 비정형 데이터 세트, 이를 포함한 데이터로 부터 분석하여 의미 있는 가치를 추출하고 결과를 분석하는 기술이다.

1-2 정형 데이터와 비정형 데이터에 대해 설명하시오.

정형 데이터(Structured data)
- 일정한 형식이나 규칙에 맞게 저장된 데이터이다.
- 정형 데이터의 예 : 스프레드시트, 관계형 데이터베이스 CSV 등

비정형 데이터(Unstructured data)
- 정해진 규칙이 없어서 값의 의미를 쉽게 파악하기 힘든 데이터들
- 비정형 데이터의 예 : SNS, 텍스트, 음성, 영상 등

의미를 분석하기 힘들었던 대용량의 비정형 데이터를 분석함으로써 새로운 인사이트(Insight)를 얻을 수 있다.

1-3 빅데이터의 처리 과정에 대해 설명하시오.
1. 데이터 소스 → 2. 수집 → 3. 저장 → 4. 처리 → 5. 분석 → 6. 표현 

위와 같이 처리 과정은 6단계로 구성된다.

1-4 빅데이터 활용 사례를 2가지 이상 설명하시오.
1. 당근마켓
- 게시한 사진의 상품 카테고리 자동 분류, 맞춤형 콘텐츠 추천

2. 넷플릭스(Netflix)
- 사용자의 영화 관람 목록에 기초해 새로운 영화를 추천해주는 시스템 개발

3. 신용카드사
- 빅데이터 분석을 통해 파악한 고객 니즈와 스마트폰으로부터 수집한 정보를 결합하는 CLO(Card Linked Offer)서비스를 마케팅에 활용

4. 페이팔(PayPal)
- 온라인 쇼핑몰 사기수법을 분석 및 예방하기 위해 딥러닝을 도입
